```python
## Step 1.3 Data Ingestion Sandbox (SAS -> DataFrame -> CSV)

import pandas as pd

raw_path_806 = "data/raw/m_d_806.sas7bdat"
raw_path_903 = "data/raw/m_d_903.sas7bdat"

df806 = pd.read_sas(raw_path_806, format="sas7bdat", encoding="latin1")
df903 = pd.read_sas(raw_path_903, format="sas7bdat", encoding="latin1")

print("df806 shape:", df806.shape)
print("df903 shape:", df903.shape)

display(df806.head())
display(df903.head())

miss_806 = df806.isna().mean().sort_values(ascending=False).head(15)
miss_903 = df903.isna().mean().sort_values(ascending=False).head(15)
print("\nTop missingness (806):\n", miss_806)
print("\nTop missingness (903):\n", miss_903)
```

```python
out_806 = "data/processed/census_806.csv"
out_903 = "data/processed/census_903.csv"

df806.to_csv(out_806, index=False)
df903.to_csv(out_903, index=False)

print("Saved:", out_806, out_903)
```
