```python
df = pd.read_parquet('../data/processed/census80_cleaned.parquet')

df.columns = [str(col).strip() for col in df.columns]

print(df.shape)
print('SEXK' in df.columns)
print('SEX2ND' in df.columns)
print('KIDCOUNT' in df.columns)
print(df.columns.tolist())
```

```python
print(df['SEXK'].head(10))
print(df['SEX2ND'].head(10))
print(df['SEXK'].value_counts(dropna=False).head(10))
print(df['SEX2ND'].value_counts(dropna=False).head(10))
```

```python
import pandas as pd

def clean_numeric(s):
    return pd.to_numeric(
        s.astype(str).str.strip(),
        errors='coerce'
    )

df['SEXK'] = clean_numeric(df['SEXK'])
df['SEX2ND'] = clean_numeric(df['SEX2ND'])
df['HOURSM'] = clean_numeric(df['HOURSM'])
df['AGEM'] = clean_numeric(df['AGEM'])
df['MARITAL'] = clean_numeric(df['MARITAL'])
df['RACEM'] = clean_numeric(df['RACEM'])
```

```python
import numpy as np

df['same_sex'] = np.where(
    df['SEXK'].notna() & df['SEX2ND'].notna(),
    (df['SEXK'] == df['SEX2ND']).astype(int),
    np.nan
)

df['more_kids'] = np.where(
    df['KIDCOUNT'].notna(),
    (df['KIDCOUNT'] >= 3).astype(int),
    np.nan
)

print(df[['SEXK', 'SEX2ND', 'same_sex', 'KIDCOUNT', 'more_kids']].head())
print(df['same_sex'].value_counts(dropna=False))
print(df['more_kids'].value_counts(dropna=False))
```

```python
import statsmodels.api as sm

controls = ['AGEM', 'MARITAL', 'RACEM']
use_cols = ['HOURSM', 'more_kids', 'same_sex'] + controls

iv_df = df[use_cols].dropna().copy()

print(iv_df.shape)
iv_df.head()
```

```python
X_first = sm.add_constant(iv_df[['same_sex'] + controls])
first_stage = sm.OLS(iv_df['more_kids'], X_first).fit(cov_type='HC1')

print(first_stage.summary())
```

```python
iv_df['more_kids_hat'] = first_stage.predict(X_first)

X_second = sm.add_constant(iv_df[['more_kids_hat'] + controls])
manual_2sls = sm.OLS(iv_df['HOURSM'], X_second).fit(cov_type='HC1')

print(manual_2sls.summary())
```

```python
y1 = iv_df.loc[iv_df['same_sex'] == 1, 'HOURSM'].mean()
y0 = iv_df.loc[iv_df['same_sex'] == 0, 'HOURSM'].mean()

d1 = iv_df.loc[iv_df['same_sex'] == 1, 'more_kids'].mean()
d0 = iv_df.loc[iv_df['same_sex'] == 0, 'more_kids'].mean()

wald = (y1 - y0) / (d1 - d0)

print("E[Y|Z=1] =", y1)
print("E[Y|Z=0] =", y0)
print("E[D|Z=1] =", d1)
print("E[D|Z=0] =", d0)
print("Wald Estimator =", wald)
```
